```python
import os
import pandas as pd
import mysql.connector

# 1. Connect (The Bridge)
conn = mysql.connector.connect(
    host='localhost',
    user='root',
    password=os.getenv('DB_PASSWORD'),  # set DB_PASSWORD env var
    database='chuka_electronics'
)

# 2. Get Data (The Messenger)
df_sales = pd.read_sql("SELECT * FROM Sales", conn)
df_products = pd.read_sql("SELECT * FROM Products", conn)

# 3. Analyze (The Manipulation)
merged_df = pd.merge(df_sales, df_products, on='product_id')
# If quantity isn't present in Sales, assume one unit per sale; otherwise coerce to numeric
if 'quantity' not in merged_df.columns:
    merged_df['quantity'] = 1
else:
    merged_df['quantity'] = pd.to_numeric(merged_df['quantity'], errors='coerce').fillna(1)
# Ensure unit_price is numeric and treat missing prices as 0
merged_df['unit_price'] = pd.to_numeric(merged_df['unit_price'], errors='coerce').fillna(0)
merged_df['transaction_total'] = merged_df['quantity'] * merged_df['unit_price']
total_revenue = merged_df['transaction_total'].sum()

# 4. Success Output
print(f"Total Revenue: KES {total_revenue:,.2f}")
conn.close()
merged_df.head()
```
