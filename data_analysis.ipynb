```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Create the data manually (This replaces the database connection)
data = {
    'product_name': ['HP Laptop', 'Samsung Charger', 'Amtec Speaker', 'Extension', 'Keyboard'],
    'unit_price': [45000, 900, 4500, 450, 3600]
}

merged_df = pd.DataFrame(data)

# 2. THE CHART CODE
plt.figure(figsize=(10, 6))
sns.set_style("white")
colors = ["#FF007A", "#FF69B4", "#FFB6C1", "#FFC0CB", "#FFE3F0"]

plt.bar(merged_df['product_name'], merged_df['unit_price'], color=colors)

# 3. Titles and Labels
plt.title('Chuka Electronics: Top Product Pricing', fontsize=16, fontweight='bold', color='#2D001D', pad=20)
plt.ylabel('Price (KSh)', fontsize=12, fontweight='600', color='#FF007A')
plt.xticks(rotation=15, ha='right', color='#2D001D')
sns.despine()

plt.show()
```
